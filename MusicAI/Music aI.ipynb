```python
# Extract tracks from the playlist
def get_tracks(playlist):
    """Extract track24 info from a Spotify playlist object"""
    tracks = playlist['tracks']['items']
    track_list = []
    for item in tracks:
        track_data = item['track']
        track_list.append({
            'track_id': track_data['id'],
            'track_name': track_data['name'],
            'artist': track_data['artists'][0]['name'],
            'popularity': track_data['popularity']
        })
    return pd.DataFrame(track_list)

# Get dataframes
tracks_2024_df = get_tracks(playlist24)
tracks_2025_df = get_tracks(playlist25)

print("2024 tracks:", tracks_2024_df.shape)
print("2025 tracks:", tracks_2025_df.shape)
```

```python
def get_audio_features_batch(track_ids):
    all_features = []
    batch_size = 100  # max Spotify allows
    num_batches = math.ceil(len(track_ids)/batch_size)

    for i in range(num_batches):
        batch_ids = track_ids[i*batch_size:(i+1)*batch_size]
        features = sp.audio_features(batch_ids)
        # Remove None values (unavailable tracks)
        features = [f for f in features if f is not None]
        all_features.extend(features)
    
    return pd.DataFrame(all_features)
```
